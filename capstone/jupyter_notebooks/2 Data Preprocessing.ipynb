```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import pandasql as sq

%matplotlib inline
train_data = pd.read_csv('../datasets/train.csv')
train_data.head()
```

```python
period_train = pd.read_csv('../datasets/periods_train.csv')
period_train.head()
```

```python
train_data_stripped = train_data[['item_id','parent_category_name','category_name','price','user_type','image_top_1','deal_probability']]
train_data_stripped.head()
```

```python
full_data_raw = train_data_stripped.join(period_train,on='item_id',how='left', lsuffix='_left', rsuffix='_right')
full_data_raw.head()
```

```python
#full_data_raw.to_csv('../datasets/input_set.csv')
only_numeric_data = full_data_raw.drop(['item_id_left','parent_category_name','category_name','user_type','item_id_right','activation_date','date_from','date_to'], axis=1)
only_numeric_data.head()
only_numeric_data = only_numeric_data.fillna(0)
only_numeric_data.to_csv('../datasets/only_numeric.csv')
```

```python
numeric_with_itemseq = train_data[['price','item_seq_number','image_top_1','deal_probability']]
numeric_with_itemseq.head()
numeric_with_itemseq = numeric_with_itemseq.fillna(0)
numeric_with_itemseq.to_csv('../datasets/itemseq.csv')
```
